```python
returns = df_cleaned.iloc[:, 1:].pct_change().dropna()
returns
```

```python
import numpy as np
weights = np.ones(456) / 456
relative_portfolio = returns.dot(weights)
relative_portfolio
```

```python
print("Mean Return (how much percentage did the portfolio gain per day on average):", relative_portfolio.mean())
print("Volatility (percentage that the portfolio fluctuates per day on average):", relative_portfolio.std())
print("Sharpe Ratio (return per unit of volatility):", relative_portfolio.mean()/relative_portfolio.std())
print("VaR 95% (maximum percentage of loss on 95% of days):", relative_portfolio.quantile(0.05))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Make dates align with returns rows
dates = pd.to_datetime(df_cleaned.loc[returns.index, "Date"].astype(str), format="%Y%m%d")

# Equal-weight benchmark
n = returns.shape[1]
equal_weights = np.ones(n) / n
equal_portfolio = returns.dot(equal_weights)
equal_cum = (1 + equal_portfolio).cumprod()
```

```python
# Minimalist model: Minimum Variance Portfolio

cov_matrix = returns.cov()

def portfolio_volatility(w):
    return np.sqrt(w.T @ cov_matrix @ w)

x0 = np.ones(n) / n
constraints = {"type": "eq", "fun": lambda w: np.sum(w) - 1}
bounds = [(0, 1)] * n

result = minimize(
    portfolio_volatility,
    x0,
    method="SLSQP",
    bounds=bounds,
    constraints=constraints
)

print("Optimization success:", result.success)
print("Message:", result.message)

min_var_weights = result.x
min_var_portfolio = returns.dot(min_var_weights)
min_var_cum = (1 + min_var_portfolio).cumprod()
```

```python
# Summary metrics 
summary = pd.DataFrame({
    "Portfolio": ["Equal Weight", "Minimalist"],
    "Mean Daily Return": [equal_portfolio.mean(), min_var_portfolio.mean()],
    "Daily Volatility": [equal_portfolio.std(), min_var_portfolio.std()],
    "Sharpe Ratio": [
        equal_portfolio.mean() / equal_portfolio.std(),
        min_var_portfolio.mean() / min_var_portfolio.std()
    ],
    "VaR 95%": [
        equal_portfolio.quantile(0.05),
        min_var_portfolio.quantile(0.05)
    ]
})

summary
```

```python
# Top 10 holdings by weight
top10 = pd.Series(min_var_weights, index=returns.columns)\
    .sort_values(ascending=False)\
    .head(10)

top10
```

```python
# Figure 5: Efficient Frontier via Monte Carlo
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n = returns.shape[1]
mc_vols, mc_rets = [], []

for _ in range(5000):
    # Randomly select a subset of 10 to 50 stocks, rather than all 456.
    k = np.random.randint(10, 51)
    chosen = np.random.choice(n, k, replace=False)
    w_sub = np.random.dirichlet(np.ones(k))
    w = np.zeros(n)
    w[chosen] = w_sub

    mc_rets.append(returns.dot(w).mean())
    mc_vols.append(returns.dot(w).std())

plt.figure(figsize=(11, 7))

# 背景随机点云
plt.scatter(mc_vols, mc_rets, alpha=0.3, s=8, c='steelblue', label="Random Portfolios")

# Minimum Variance 点
plt.scatter(
    min_var_portfolio.std(),
    min_var_portfolio.mean(),
    color='red', s=150, zorder=5, label="Minimum Variance"
)

# Equal Weight 点
plt.scatter(
    equal_portfolio.std(),
    equal_portfolio.mean(),
    color='orange', s=150, zorder=5, label="Equal Weight"
)

# label
plt.annotate(
    "Minimum Variance",
    (min_var_portfolio.std(), min_var_portfolio.mean()),
    textcoords="offset points", xytext=(10, 6), fontsize=10, color='red'
)
plt.annotate(
    "Equal Weight",
    (equal_portfolio.std(), equal_portfolio.mean()),
    textcoords="offset points", xytext=(10, -12), fontsize=10, color='orange'
)

plt.xlabel("Daily Volatility (Risk)", fontsize=12)
plt.ylabel("Mean Daily Return", fontsize=12)
plt.title("Efficient Frontier - Monte Carlo Simulation (5000 Portfolios)", fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('fig5_efficient_frontier.png', transparent=True, dpi=150)
plt.show()
```
