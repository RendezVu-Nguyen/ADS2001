# Keep stocks with 90% coverage

```python
coverage = df.notna().mean()
df = df.loc[:, coverage >= 0.9]
df
```

# Drop any rows with N/A values

```python
df_clean = df.dropna()
df_clean
```

# Remove duplicated rows (weekends and public holidays) and non-publicly traded entities (numerical tickers)

```python
df_cleaned = df_clean[~df_clean.iloc[:, 1:].duplicated()]
df_cleaned = df_cleaned.drop(df.columns[1], axis=1)
df_cleaned
```

# Convert raw prices into returns

```python
returns = df_cleaned.iloc[:, 1:].pct_change().dropna()
returns
```

# Apply static weight (relative) portfolio on every stock equally

```python
import numpy as np
weights = np.ones(456) / 456
relative_portfolio = returns.dot(weights)
relative_portfolio
```

# Calculate mean return, volatility, variance, sharpe ratio, and VaR 95%

```python
print("Mean Return (how much percentage did the portfolio gain per day on average):", relative_portfolio.mean())
print("Volatility (percentage that the portfolio fluctuates per day on average):", relative_portfolio.std())
print("Sharpe Ratio (return per unit of volatility):", relative_portfolio.mean()/relative_portfolio.std())
print("VaR 95% (maximum percentage of loss on 95% of days):", relative_portfolio.quantile(0.05))
```

# Apply static weight (relative) portfolio on every stock equally and get sortino value

```python
equal_weights = np.ones(456) / 456
equal_port = returns.dot(equal_weights)

mean_eq = equal_port.mean()
downside_eq = np.minimum(0, equal_port)
downside_std_eq = downside_eq.std()

sortino_eq = mean_eq / downside_std_eq
print(sortino_eq)
```

# Generate random weights

```python
import numpy as np

weights_val = []
sortino_val = []
returns_val = []
downside_vol_val = []

for i in range(10000):
    weights = np.random.uniform(0, 1, 456)
    proportions = weights / weights.sum()
    
    weights_val.append(proportions)
    
    ran_port = returns.dot(proportions)
    
    meanvalue = ran_port.mean()
    
    downside = np.minimum(0, ran_port)
    downside_std = downside.std()
    
    returns_val.append(meanvalue)
    downside_vol_val.append(downside_std)
    sortino_val.append(meanvalue / downside_std)
    
max_sortino = max(sortino_val)
ind = sortino_val.index(max_sortino)
best_weights = weights_val[ind]
print(max_sortino, ind)
print(best_weights)
```

# Dynamic Portfolio

```python
window = 30
dynamic_returns = []

for t in range(window, len(returns)):
    past_returns = returns.iloc[t-window:t]
    
    # average past performance
    scores = past_returns.mean()
    
    # make all values positive
    scores = scores - scores.min()
    
    # normalize to weights
    weights = scores / scores.sum()
    
    # apply to current day
    today_return = returns.iloc[t].dot(weights)
    
    dynamic_returns.append(today_return)
    
dynamic_returns = np.array(dynamic_returns)

mean_dyn = dynamic_returns.mean()
downside_dyn = np.minimum(0, dynamic_returns)
downside_std_dyn = downside_dyn.std()

sortino_dyn = mean_dyn / downside_std_dyn

print(sortino_dyn)
```
